```python
import numpy as np
df["MILES_CAT*"]= np.where(df["MILES*"] >10 , "Long trip", "short trip")
df
```

```python
# Adding a new column based on the condition using np.where
df["MILES_CAT*"] = np.where(df["MILES*"] > 20, "Long trip", 
                            np.where(df["MILES*"] > 10, "Medium trip", "Short trip"))
df
```
